```python
diabetes = pd.read_csv("diabetes.csv")
df=diabetes.copy()
df=df.dropna()
df.head()
y=df["Outcome"]
X=df.drop(["Outcome"], axis = 1)
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=42)
```

 *  Modelimizin test train olarak ayarladık ilk parf da çektik sonra bi kopya her ihtimale karşı boşları dropladık predict parametremizi belirledik ve başa aldık kodun düzgünlüğü tamamen görsel değil tabi ardından ilk başta söz ettiğim mevzuyu yaptım *

```python
scaler = StandardScaler()
```

standartlaştırma işlemi gerekiyo etkilenmeyi azaltmak için

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

girdilerimizi scale ediyoruz mlp modeline öyle gerekiyo weightler falan belli bi aralıkta olması gerekiyo bcs

```python
X_train_scaled[0:5]
```

```python
mlpc = MLPClassifier().fit(X_train_scaled, y_train)
```

```python
y_pred = mlpc.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```
